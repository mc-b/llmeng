## OpenAI-API

Die OpenAI-API bietet eine einfache Schnittstelle zu modernsten KI- Modellen für Textgenerierung, Verarbeitung natürlicher Sprache, Computer Vision und mehr. 

Erstellt einen API-Schlüssel und führt den ersten API-Aufruf durch, um loszulegen. Entdeckt, wie man Texte generiert, Bilder analysiert, Agenten erstellt und vieles mehr.

* [Erstellen eines API Schlüssel](https://platform.openai.com/api-keys)

Als nächstes Setzen wir den API-Schlüssel in `~/work/env.py`für die weiteren Übungen

```python
%%bash
source ~/work/env.py
cat <<EOF | tee ~/work/env.py
OPENAI_API_KEY="$OPENAI_API_KEY"
HF_TOKEN=""
AI_KUBECONFIG="$AI_KUBECONFIG"
AI_MODEL="gpt-5.4"
AI_NAME=""
AI_IP="${AI_IP}"
AI_BASE_URL='https://api.openai.com/v1'
EOF
```

Für die nachfolgenden Übungen verwenden wir das offizielle [OpenAI SDK](https://github.com/openai/openai-python).

```python
%run ~/work/env.py
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY, base_url=AI_BASE_URL)

response = client.responses.create(
    model=AI_MODEL,
    input="Write a one-sentence bedtime story about a unicorn."
)

print(response.output_text)
```
