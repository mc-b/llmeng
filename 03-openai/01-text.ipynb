## Textgenerierung
Lerne, wie du ein Modell dazu anregst, Text zu generieren.

Mit der OpenAI-API können Sie ein großes Sprachmodell verwenden , um anhand einer Eingabeaufforderung Text zu generieren, ähnlich wie bei ChatGPT . Die Modelle können nahezu jede Art von Textantwort erzeugen – beispielsweise Code, mathematische Gleichungen, strukturierte JSON-Daten oder menschenähnliche Prosa.

Hier ist ein einfaches Beispiel mit der Responses API , unserer empfohlenen API für alle neuen Projekte.

```python
%run ~/data/env.py
```

```python
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY, base_url=AI_BASE_URL)

response = client.responses.create(
    model=AI_MODEL,
    input="Write a one-sentence bedtime story about a unicorn."
)

print(response.output_text)
```

### Nachrichtenrollen und Anweisungen 

Sie können dem Modell mithilfe des API-Parameters und der Nachrichtenrollen Anweisungen mit unterschiedlichen Berechtigungsstufen geben .instructions

Der instructionsParameter gibt dem Modell übergeordnete Anweisungen für sein Verhalten bei der Generierung einer Antwort, einschließlich Tonfall, Zielsetzungen und Beispielen für korrekte Antworten. Alle so übermittelten Anweisungen haben Vorrang vor Eingabeaufforderungen im inputParameter.

```python
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY, base_url=AI_BASE_URL)

response = client.responses.create(
    model=AI_MODEL,
    reasoning={"effort": "low"},
    input=[
        {
            "role": "developer",
            "content": "Talk like a pirate."
        },
        {
            "role": "user",
            "content": "Are semicolons optional in JavaScript?"
        }
    ]
)

print(response.output_text)
```

### Nachrichtenrolle in den Anweisungen

Nachrichtenrollen können auch im Prompt mitgegeben werden.

```python
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY, base_url=AI_BASE_URL)

response = client.responses.create(
    model=AI_MODEL,
    input="Schreibe ein kurzes Python Beispiel für einen REST Client."
)

print(response.output_text)
```
