### Nachrichtenrollen und Anweisungen 

Mit der Responses API können Sie dem Modell über verschiedene Nachrichtenrollen (`developer`, `user`, `assistant`, `tool`) Anweisungen und Kontext mit unterschiedlichen Prioritäten übergeben.

Die wichtigste Rolle ist dabei meistens `developer`. Diese definiert das grundlegende Verhalten des Modells, beispielsweise:

* Tonalität
* Zielsetzung
* Firmenkontext
* Antwortregeln
* Formatvorgaben
* Beispiele für gewünschte Antworten

Die eigentliche Benutzeranfrage wird anschliessend als `user`-Nachricht übergeben.

Anweisungen in der Rolle `developer` haben grundsätzlich höhere Priorität als Inhalte aus `user`-Nachrichten.

```python
from openai import OpenAI

client = OpenAI(
    api_key=OPENAI_API_KEY,
    base_url=AI_BASE_URL
)

response = client.responses.create(
    model=AI_MODEL,
    reasoning={"effort": "low"},
    input=[
        {
            "role": "developer",
            "content": """
You are an AI assistant for AutoShop GmbH.

Your tasks:
- Answer customer questions about cars, availability, services, pricing, and appointments.
- Be professional, concise, and helpful.
- Respond in German unless the customer writes in another language.
- If information is missing, ask follow-up questions.
- Never invent vehicle stock or prices.
"""
        },
        {
            "role": "user",
            "content": "Welche Automarken wurden im Jahr 2020 am meisten verkauft?"
        }
    ]
)

print(response.output_text)
```
