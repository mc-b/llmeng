```python
import os
from openai import OpenAI
from IPython.display import Markdown, display
from datetime import datetime
import requests
import base64
```

```python
client = OpenAI(
    base_url="http://10.3.24.11:11434/v1",
    api_key="ollama"  # Dummy, wird ignoriert
)

# 1. Bild herunterladen
image_url = "https://raw.githubusercontent.com/mc-b/lernvirt/8aa088ca003613c7de0872ff8137c889325e94c2/examples/aiaas/mlg/02-iris.png"
response = requests.get(image_url)
response.raise_for_status()

# 2. In Base64 umwandeln
image_bytes = response.content
image_base64 = base64.b64encode(image_bytes).decode("utf-8")

# MIME-Type bestimmen (hier JPEG, da URL .jpg ist)
mime_type = "image/png"
data_url = f"data:{mime_type};base64,{image_base64}"

# 3. An OpenAI übergeben
api_response = client.responses.create(
    model="gemma3:12b",
    input=[
        {
            "role": "user",
            "content": [
                {
                    "type": "input_text",
                    "text": "Was für Planzen sind auf diesem Bild."
                },
                {
                    "type": "input_image",
                    "image_url": data_url
                }
            ]
        }
    ]
)

print(api_response.output_text)
```

```python
# Präzisere Formulierung
api_response = client.responses.create(
    model="gemma3:12b",
    input=[
        {
            "role": "user",
            "content": [
                {
                    "type": "input_text",
                    "text": "Was für Iris Blüten sind auf diesem Bild, versuche die Unterschiede Anhand der Kronblätter (Petalen) herauszufinden"
                },
                {
                    "type": "input_image",
                    "image_url": data_url
                }
            ]
        }
    ]
)

print(api_response.output_text)
```
