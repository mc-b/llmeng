# Model Context Protocol 

ist ein offenes Protokoll, mit dem Tools, Dateien, Repositories, Datenbanken oder APIs strukturiert an ein LLM angebunden werden.

Ollama liefert dabei nur die Text-Generierung; Kontext, Tools und Ressourcen kommen über MCP von aussen.

**Architektur**
* Ollama läuft auf einem separaten Rechner ($AI_BASE_URL)
* Ein MCP-Client spricht MCP nach vorne
* Derselbe Client spricht nach hinten mit Ollama (OpenAI-kompatible API)
* MCP-Server stellen Ressourcen bereit (Files, Git, SQL, HTTP, …)

```python
from mcp.server.fastmcp import FastMCP

mcp = FastMCP("webshop")

PRODUCTS = {
    "1": {"name": "Tastatur", "price": 79.0, "stock": 5},
    "2": {"name": "Maus", "price": 39.0, "stock": 12},
}

@mcp.resource("products://")
def list_products():
    return [{"id": k, **v} for k, v in PRODUCTS.items()]

@mcp.tool()
def buy_product(id: str, quantity: int):
    PRODUCTS[id]["stock"] -= quantity
    return "ok"

# main
await mcp.run_sse_async()
```
